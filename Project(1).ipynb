### 포물선 궤적 그리기

```python
from matplotlib import pyplot as plt
import math
```

```python
def parabolic_graph(x,y):
    plt.plot(x,y)
    plt.xlabel('x_coordinate')
    plt.ylabel('y_coordinate')
    plt.title('Parabolic Motion')
    
def flight_coordinate(start, final, interval):
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
    return numbers
```

```python
def parabolic_trajectory(u, theta):
    # u is intial velocity
    theta = math.radians(theta)
    g = 9.80665
    
    # checking time interval
    t_flight = 2*u*math.sin(theta)/g
    intervals = flight_coordinate(0, t_flight, 0.01)
    
    x = []
    y = []
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t-0.5*g*(t**2))
        
    parabolic_graph(x,y)
```

```python
if __name__ == '__main__':
    try:
        u = float(input('what is the initial velocity?'))
        theta = float(input('what is the angle of projection(degrees)?'))
    except ValueError:
        print('try again!')
    else:
        parabolic_trajectory(u,theta)
        plt.show
```
